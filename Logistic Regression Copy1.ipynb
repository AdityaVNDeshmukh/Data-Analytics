```python
df.drop(columns = ['pdays', 'previous'], inplace = True)
```

```python
df.columns
```

```python
# Removing the outliers
for i in df.columns:
    if df[i].dtype != 'object':
        Q1 = df[i].quantile(0.25)
        Q3 = df[i].quantile(0.75)
        IQR = Q3 - Q1
        LL = Q1 - (1.5*IQR)
        UL = Q3 + (1.5*IQR)
        df[i] =  df[i][(df[i]>=LL) & (df[i]<=UL)]  # Here we have written df[i] because, we are not considering all the columns from df. 
```

```python
for i in df.columns:
    if df[i].dtype != 'object':
        sns.boxplot(df[i])
        plt.show()
```

```python
# After removing the outliers, we might get to see a lot of null values. So lets check
df.isna().sum()
```

```python
df.dropna(inplace = True)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 85)
```

```python
X_train
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
# Comparing actual values with predicted values.
list(zip(Y_test, Y_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test, Y_pred)
```

```python
# Using the above matrix, we can calculate the accuracy of a model.
# Step 1:- add a11 and a22 (a11 + a22)
# Step 2:- add a11, a12, a21, a22 (a11 + a12 + a21 + a22)
# Step 3:- accuracy_score = (a11 + a22) / (a11 + a12 + a21 + a22)----->(6373 + 101)/(6373 + 55 + 707 + 101) = 6474/7236 = 0.894693
```

```python
from  sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, Y_pred))
```
