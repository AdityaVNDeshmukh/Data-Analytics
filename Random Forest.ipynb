```python
df.drop(columns = 'customerID', axis = 1, inplace = True)
```

```python
#Label encoding
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = le.fit_transform(df[i])
```

```python
df.head(10)
```

```python
df.info()
```

```python
# Feature Selection ----- Chi square test ------SelectKBest
```

```python
from sklearn.feature_selection import chi2, SelectKBest
```

```python
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]
```

```python
X.head(2)
```

```python
Y.head(2)
```

```python
# We are selecting the K Best features with the help of chi2
# K = no.of features
# Model = chi2
selector = SelectKBest(chi2, k = 8)
X_new = selector.fit_transform(X, Y)
```

```python
X_new
```

```python
selected_features = X.columns[selector.get_support()]
selected_features
```

```python
X_update = df[selected_features]
X_update
```

```python
# Now we have both the variables i.e. X and Y
# X ----> X_update
# Y ----> Y
```

```python
# Model Building
```

```python
# Now split the data into training and testing parts  -----> 80:20
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X_update, Y, test_size = 0.2, random_state = 55)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# Calling Random Forest
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```

```python
# To enhance the accuracy score and redure over fitting, we can use
# gridsearchcv and randomsearchcv
```

```python
# randomsearchcv
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# n_estimators -----> How many decision Trees we want to create  ----> min 5
# max_depth ----> How deep you want to go in each of the decision tree
# min_samples_leaf ----> No. of min. sample leafs to be present in each DT
# min_samples_split ----> No. of min splits
# max_features ----> No. of maximum features -----> to reduce over fitting ----> sqrt, log2, auto
```

```python
# randint
```

```python
from scipy.stats import randint
```

```python
parameters = {'n_estimators': randint(5, 200),
             'max_depth': randint(5, 30),
             'min_samples_leaf': randint(2, 14),
             'min_samples_split': randint(1, 5),
             'max_features': ['sqrt', 'log2', 'auto']}
```

```python
random_search = RandomizedSearchCV(rf, param_distributions = parameters, n_iter = 100, cv = 5, n_jobs = -1)
random_search.fit(x_train, y_train)
print(random_search.best_params_)
print(random_search.best_score_)
```
