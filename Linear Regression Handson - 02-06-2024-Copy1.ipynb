```python
# Creating a code to get boxplots for all the columns having numeric dtypes.
for i in cols:
    if(df[i].dtypes == 'int64' or df[i].dtypes == 'float64'):
        plt.boxplot(df[i])
        plt.xlabel(i)
        plt.ylabel('Outliers')
        plt.show()
```

```python
# Removal of outliers using IQR method
```

```python
# For bmi column
Q1 = df.bmi.quantile(0.25)
Q3 = df.bmi.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.bmi >= Q1 - 1.5 * (IQR)) & (df.bmi <= Q3 + 1.5 * (IQR))]

# For past_consultations column
Q1 = df.past_consultations.quantile(0.25)
Q3 = df.past_consultations.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.past_consultations >= Q1 - 1.5 * (IQR)) & (df.past_consultations <= Q3 + 1.5 * (IQR))]

# For Hospital_expenditure column
Q1 = df.Hospital_expenditure.quantile(0.25)
Q3 = df.Hospital_expenditure.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Hospital_expenditure >= Q1 - 1.5 * (IQR)) & (df.Hospital_expenditure <= Q3 + 1.5 * (IQR))]

# For NUmber_of_past_hospitalizations column
Q1 = df.NUmber_of_past_hospitalizations.quantile(0.25)
Q3 = df.NUmber_of_past_hospitalizations.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.NUmber_of_past_hospitalizations >= Q1 - 1.5 * (IQR)) & (df.NUmber_of_past_hospitalizations <= Q3 + 1.5 * (IQR))]

# For Anual_Salary column
Q1 = df.Anual_Salary.quantile(0.25)
Q3 = df.Anual_Salary.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.Anual_Salary >= Q1 - 1.5 * (IQR)) & (df.Anual_Salary <= Q3 + 1.5 * (IQR))]

# For charges column
Q1 = df.charges.quantile(0.25)
Q3 = df.charges.quantile(0.75)

IQR = Q3 - Q1

df = df[(df.charges >= Q1 - 1.5 * (IQR)) & (df.charges <= Q3 + 1.5 * (IQR))]
```

```python
# Model Building
```

```python
# To locate the target column and store it in y as dependent variable.

y = df.iloc[:,-1]
```

```python
# To store all the features other than target in x as independent variables.

x = df.loc[:,['children', 'Claim_Amount',
       'past_consultations', 'Hospital_expenditure', 'age']]
```

```python
# Splitting the variables i.e. x & y
# The variables x & y are split into 2 different categories making it 4 variables in the end.
# 4 variables:- x_train, x_test, y_train, y_test

# Training data is for actual values
# Testing data is for predicted values.
```

```python
from sklearn.model_selection import train_test_split         # Library for splitting
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 50)
```

```python
# import the algorithm of linear regression
from sklearn.linear_model import LinearRegression
```

```python
# Created a model and stored it in the variable lr
lr = LinearRegression()
lr
```

```python
# .fit() method is used to provide the training data as input in the machine.
lr.fit(x_train, y_train)
```

```python
# Using the test variables (x_test, y_test) perform predictions of result i.e. predicted outcomes.
```

```python
# Crate a variable and make use of .predict() method with x_test as an input in it.
# This gives us the output values that are predicted outputs that are received from the training data
```

```python
pred = lr.predict(x_test)
pred
```

```python
# Compare with y_test for better understanding of accuracy of the model built for the comparison of expected outputs and predicted outputs.
# IN short calculate r2s (r2_score)
```

```python
from sklearn.metrics import *
```

```python
score = r2_score(y_test, pred)
score
```
